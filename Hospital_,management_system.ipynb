```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

# Load the dataset

```python
df = pd.read_csv('/content/patients_500.csv')
df
```

# EDA

```python
df.columns
```

## now copy the datasets

```python
df_copy = df.copy()
```

## calculate the bmi and add this features.

```python
# features - 01
df_copy['bmi']= df_copy['weight']/(df_copy['height']/100)**2
df_copy
```

## instead of using directly age we create a new feature called age_group.

```python
# function to calculate the age avarage
def age_avarage(age):
  if age < 18:
    return 'child'
  elif age >= 18 and age <= 65:
    return 'adult'
  else:
    return 'senior'
```

```python
# features - 02
df_copy['age_group'] = df_copy['age'].apply(age_avarage)
df_copy
```

```python

# features - 03
df_copy["smoking"] = [random.choices(
    ["Non-smoker", "Occasional", "Regular"],
    weights=[0.6, 0.25, 0.15]
)[0] for _ in range(len(df))]
```

```python
df_copy
```

```python
#lifecycle risk function.
def lifecycle_risk(row):
    if row['smoking'] == "Regular" and row['bmi'] >= 30:
        return "High Risk"

    elif row['smoking'] == "Occasional" and row['bmi'] >= 25:
        return "Medium Risk"

    else:
        return "Low Risk"
```

```python
# features - 04 : lifecycle risk
df_copy['life_cycle']=df_copy.apply(lifecycle_risk,axis=1)
df_copy
```

# Add new features city. and add the random city name.

```python

cities = [
    "Dhaka", "Chittagong", "Sylhet", "Rajshahi",
    "Khulna", "Barisal", "Rangpur", "Mymensingh",
    "Comilla", "Narayanganj", "Gazipur", "Jessore",
    "Bogura", "Dinajpur", "Noakhali"
]
```

```python
# features - 05 : add city in the dataset
```

```python
df_copy = df_copy.assign(city=np.random.choice(cities, len(df_copy)))
df_copy
```

# ekhon amra randomly city er nam use na kore tire create korbo.

```python
tire1_city = ["Dhaka", "Khulna", "bogura", "Rangpur"]
tier2_city = [
    "Chittagong",
    "Sylhet",
    "Rajshahi",
    "Barisal",
    "Mymensingh",
    "Comilla",
    "Narayanganj",
    "Gazipur",
    "Jessore",
    "Dinajpur",
    "Noakhali",
    "Cox's Bazar",
    "Pabna",
    "Tangail",
    "Feni",
    "Brahmanbaria",
    "Kushtia",
    "Sirajganj",
    "Jamalpur",
    "Satkhira"
]
```

# write the function for city tier.

```python
def city_tier(city):
    if city in tire1_city:
        return 1
    elif city in tier2_city:
        return 2
    return 3
```

# now add the features on the datasets

```python
df_copy = df_copy.assign(city_tier=df_copy['city'].apply(city_tier))
df_copy
```

```python
# features - 06: Add income_lpa
incomes = [
    "Below 3 LPA",
    "3-6 LPA",
    "6-10 LPA",
    "10-15 LPA",
    "Above 15 LPA"
]
df_copy["income_lpa"] = np.random.choice(incomes, len(df_copy))



df_copy
```

```python
# features - 07: Add occupations
occupations = [
    "Student",
    "Employed",
    "Self-Employed",
    "Unemployed",
    "Retired"
]
df_copy["occupations"] = np.random.choice(occupations, len(df_copy))
df_copy
```

```python
# features - 08: Add insurence_premium_catagory
premium_categories = [
    "Low",
    "Medium",
    "High",
    "Very High"
]
df_copy["insurence_premium_catagory"] = np.random.choice(premium_categories, len(df_copy))
df_copy
```

# Now we target our features to tain and the further work.

```python
X = df_copy[['bmi', 'age_group', 'life_cycle', 'city_tier', 'occupations', 'income_lpa']]
Y = df_copy['insurence_premium_catagory']
```

## catagorical and numaric features.

```python
catagorical_features = ['age_group', 'life_cycle', 'city_tier', 'occupations', 'income_lpa']
numarical_features = ['bmi']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
preprocessor = ColumnTransformer(
    transformers = [
        ("categorical", OneHotEncoder(sparse_output=False), catagorical_features),
        ("numerical", StandardScaler(), numarical_features)
    ]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

```

```python
pipelines = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier())
])
```

```python
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

```

```python
# Fit the pipeline
pipelines.fit(X_train, Y_train)

print("Pipeline created and fitted successfully!")
```

```python
accuracy = pipelines.score(X_test, Y_test)
print(f'Test Accuracy: {accuracy:.2%}')
```

```python
# Evaluate the model
accuracy = pipelines.score(X_test, Y_test)
print(f"Model Accuracy: {accuracy:.2%}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search
param_grid = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__criterion': ['gini', 'entropy']
}

# Initialize GridSearchCV
grid_search = GridSearchCV(pipelines, param_grid, cv=5, n_jobs=-1, verbose=1)

# Fit the grid search
grid_search.fit(X_train, Y_train)

# Best parameters and score
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best Cross-validation Accuracy: {grid_search.best_score_:.2%}")

# Evaluate the best model on test set
best_accuracy = grid_search.score(X_test, Y_test)
print(f"Tuned Model Test Accuracy: {best_accuracy:.2%}")
```
